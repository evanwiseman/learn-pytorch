```python
import torch
import numpy as np
```

```python
# From list
data = [[1, 2], [3, 4]]
x_data = torch.tensor(data)
print(x_data)
```

```python
# From np array
np_array = np.array(data)
x_np = torch.from_numpy(np_array)
print(x_np)
```

```python
# From another Tensor
x_ones = torch.ones_like(x_data)
print(f"Ones Tensor:\n{x_ones}\n")

x_rand = torch.rand_like(x_data, dtype=torch.float)
print(f"Rand Tensor:\n{x_rand}\n")
```

```python
# From shape
shape = (2,3,)
rand_tensor = torch.rand(shape)
ones_tensor = torch.ones(shape)
zeros_tensor = torch.zeros(shape)
print(f"Random Tensor:\n{rand_tensor}\n")
print(f"Ones Tensor:\n{ones_tensor}\n")
print(f"Zeros Tensor:\n{zeros_tensor}\n")
```

```python
# Attributes of Tensor
tensor = torch.rand(3, 4)

print(f"Shape of tensor: {tensor.shape}")
print(f"Datatype of tensor: {tensor.dtype}")
print(f"Device tensor is stored on: {tensor.device}")
```

```python
# Operations on Tensors
if torch.accelerator.is_available():
    print("Moved to accelerator")
    tensor = tensor.to(torch.accelerator.current_accelerator())

print(f"Shape of tensor: {tensor.shape}")
print(f"Datatype of tensor: {tensor.dtype}")
print(f"Device tensor is stored on: {tensor.device}")
```

```python
# Standard numpy like indexing and slicing
tensor = torch.ones(4, 4)

print(f"First row: {tensor[0]}")
print(f"First column: {tensor[:, 0]}")
print(f"Last column: {tensor[..., -1]}")
tensor[:,-1] = 0
print(tensor)
```

```python
# Joining tensors
t1 = torch.cat([tensor, tensor, tensor], dim=1)
print(t1)
```

```python
# Arithmetic operations

# This computes the matrix multiplication between two tensors. y1, y2, y3 will have the same value
# ``tensor.T`` returns the transpose of a tensor
y1 = tensor @ tensor.T
y2 = tensor.matmul(tensor.T)
y3 = torch.rand_like(y1)
torch.matmul(tensor, tensor.T, out=y3)
print(f"y1 tensor:\n{y1}\n")
print(f"y2 tensor:\n{y2}\n")
print(f"y3 tensor:\n{y3}\n")

# This computes the element-wise product. z1, z2, z3 will have the same value
z1 = tensor * tensor
z2 = tensor.mul(tensor)
z3 = torch.rand_like(tensor)
torch.mul(tensor, tensor, out=z3)
print(f"z1 tensor:\n{z1}\n")
print(f"z2 tensor:\n{z2}\n")
print(f"z3 tensor:\n{z3}\n")

```

```python
# Single element tensors
agg = tensor.sum()
agg_item = agg.item()
print(agg_item, type(agg_item))
```

```python
# In-place operations
print(f"{tensor}\n")
tensor.add_(5)
print(tensor)
```

```python
# Tensor to numpy
t = torch.ones(5)
n = t.numpy()
print(f"t: {t}")
print(f"n: {n}")

# Changes to tensor array reflect in numpy
t.add_(1)
print(f"t: {t}")
print(f"n: {n}")
```

```python
# Numpy to tensor
n = np.ones(5)
t = torch.from_numpy(n)
print(f"t: {t}")
print(f"n: {n}")

# Changes to numpy array reflect in tensor
np.add(n, 1, out=n)
print(f"t: {t}")
print(f"n: {n}")
```
